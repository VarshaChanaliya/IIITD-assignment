```python
import pandas as pd
from geopy.distance import geodesic
```

```python
data = pd.read_csv("addresses.csv")
len(data)
```

```python
base_address=(13.031608,77.6259342)
threshold_distance=2
distance = []
```

```python
for i in range(0,len(data)):
    data_address = data.coordinates[i]
    value = geodesic(data_address,base_address).kilometers
    distance.insert(i,value)
    
```

```python
data.loc[:,'distance_from_base_addrs'] = distance
list = []
for x in range (0,len(data)):
    if (data.distance_from_base_addrs[x]<2):
        list.insert(x,'distance less than 2kms')
    else: list.insert(x,'distance greater than 2kms')

data.loc[:,'label'] = list
data = data.set_index("label")
data = data.drop("distance less than 2kms", axis=0)
```
